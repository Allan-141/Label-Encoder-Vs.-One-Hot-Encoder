## --------------------------------------------1-LabelEncoder------------------------------------------------

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("survey_results_public.csv")
```

```python
df.head()
```

```python
df["EdLevel"].unique()
```

```python
df["MainBranch"].unique()
```

```python
df['Hobbyist'].unique()
```

```python
df['NEWLearn'].unique()
```

### Call LabelEncoder

```python
lbl_encode = LabelEncoder()
```

### Pass the column to transform from categorical to numeric

```python
lbl_encode.fit_transform(df['NEWLearn'])
```

### Replace the converted column in the dataframe

```python
df['NEWLearn_Label_Encoded'] = lbl_encode.fit_transform(df['NEWLearn'])
```

```python
df
```

### Counting the values in the converted column

```python
df['NEWLearn_Label_Encoded'].value_counts()
```

### confirm if they are the same before and after conversion

```python
df['NEWLearn'].value_counts()
```
